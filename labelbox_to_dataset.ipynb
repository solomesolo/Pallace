```python
import pandas as pd
import os
import urllib.request
from tqdm import tqdm
tqdm.pandas()
```

```python
DATASET_FOLDER = '../data/input/Pallace_data/'

LABELS_FOLDER = DATASET_FOLDER+'exported_labels/'
DATA_FOLDER = DATASET_FOLDER+'images/'
DATASET_CSV = DATASET_FOLDER+'dataset_path_labels.csv'


LIST_LABELS_PATH = [LABELS_FOLDER + csv for csv in os.listdir(LABELS_FOLDER)]
```

```python
list_csvs = []
for csv_filename in LIST_LABELS_PATH:
    df = pd.read_csv(csv_filename, index_col=None, header=0)
    list_csvs.append(df)

df_labelbox = pd.concat(list_csvs, axis=0, ignore_index=True)
print("df.shape:", df_labelbox.shape)
df_labelbox.head()
```

# Preprocess dataframe

```python
df_labelbox['Label'] = df_labelbox['Label'].apply(lambda x: x if x!='{}' else None)
print('0 labels number:', df_labelbox[df_labelbox['Label'].isna()].shape)

df_labelbox['Labeled Data'] = df_labelbox['Labeled Data'].str.replace('&amp;', '&')
df_labelbox.head()
```

# Download images

```python
# urllib.request.urlretrieve("http://www.gunnerkrigg.com//comics/00000001.jpg", "00000001.jpg")

def download_image(raw_link):
    print("raw_link:", raw_link)
    link = raw_link.replace('&amp;', '&')
    print("link:", link)
    urllib.request.urlretrieve(link, "00000001.jpg")
    
    
```

```python
def prepare_dataframe(DATASET_CSV):
    try:
        dataframe = pd.read_csv(DATASET_CSV)
    except:
        dataframe = pd.DataFrame(columns=['FilePath', 'Label'])
    return dataframe
    
def get_max_folder_id(path):
    list_folders = os.listdir(DATA_FOLDER)
    list_folders = [int(folder.split('_')[0]) for folder in list_folders]
    return max([0]+[int(folder) for folder in list_folders])
    
def get_image_and_label(row):
    raw_link = str(row['Labeled Data'])
    link = raw_link.replace("&amp;", "&")
    
    label = row.Label
    label = 0 if label is None else 1
    
    # create folder and prepare image path
    directory = DATA_FOLDER + str(get_max_folder_id(DATA_FOLDER)+1) +'_'+str(label)
    if not os.path.exists(directory):
        os.makedirs(directory)
    path_to_save = directory + '/' + link.split('?')[0].split('%2F')[1]
    # download and save image
    urllib.request.urlretrieve(link, path_to_save)

    return path_to_save, label

dataframe = prepare_dataframe(DATASET_CSV)

# iterate df_labelbox and download images
dataframe_newpart = pd.DataFrame(columns=['FilePath', 'Label'])
for index, row in tqdm(df_labelbox.iterrows()):
    path, label = get_image_and_label(row)
    dataframe_newpart = dataframe_newpart.append(pd.DataFrame({'FilePath': path, 'Label': label}, index=[1]))

dataframe_newpart.reset_index(drop=True, inplace=True)

print("dataframe shape before new part export:", dataframe.shape)
dataframe = dataframe.append(dataframe_newpart)
print("dataframe shape before after adding new data:", dataframe.shape)


```

```python
broken_images_index = [13, 152, 275]
dataframe = dataframe.drop(broken_images_index).reset_index(drop=True)
print("dataframe.shape:", dataframe.shape)
```

```python
dataframe.to_csv(DATASET_CSV)
```
